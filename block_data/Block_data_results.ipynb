###III- 2/ Stats test 

```
import pandas as pd
dataframe = pd.DataFrame()

dataframe['State'] = ['Lent']*(len(list_alphas[2])*len(list_alphas[2]['C4']))+['Rapide']*len(list_alphas[3])*(len(list_alphas[3]['P4']))+['Libre']*(len(list_alphas[4]))*(len(list_alphas[4]['O2']))
dataframe['alpha'] = list_alphas[2]['C4'] + list_alphas[2]['P4'] + list_alphas[2]['O2']+ list_alphas[2]['Fp'] +list_alphas[3]['C4'] + list_alphas[3]['P4'] + list_alphas[3]['O2']+ list_alphas[3]['Fp']+ list_alphas[4]['C4'] + list_alphas[4]['P4'] + list_alphas[4]['O2'] + list_alphas[4]['Fp']
dataframe['beta'] = list_betas[2]['C4'] + list_betas[2]['P4'] + list_betas[2]['O2'] + list_betas[2]['Fp'] +list_betas[3]['C4'] + list_betas[3]['P4'] + list_betas[3]['O2']+ list_betas[3]['Fp'] + list_betas[4]['C4'] + list_betas[4]['P4'] + list_betas[4]['O2'] + list_betas[4]['Fp']
dataframe['theta'] = list_thetas[2]['C4'] + list_thetas[2]['P4'] + list_thetas[2]['O2']+ list_thetas[2]['Fp'] + list_thetas[3]['C4'] + list_thetas[3]['P4'] + list_thetas[3]['O2']+ list_thetas[3]['Fp'] + list_thetas[4]['C4'] + list_thetas[4]['P4'] + list_thetas[4]['O2'] + list_thetas[4]['Fp']
dataframe['baratio'] = list_ab[2]['C4'] + list_ab[2]['P4'] + list_ab[2]['O2']+ list_ab[2]['Fp'] + list_ab[3]['C4'] + list_ab[3]['P4'] + list_ab[3]['O2']+ list_ab[3]['Fp'] + list_ab[4]['C4'] + list_ab[4]['P4'] + list_ab[4]['O2'] + list_ab[4]['Fp']
#dataframe['delta'] = list_deltas[2]['C4'] + list_deltas[2]['P4'] + list_deltas[2]['O2'] + list_deltas[3]['C4'] + list_deltas[3]['P4'] + list_deltas[3]['O2'] + list_deltas[4]['C4'] + list_deltas[4]['P4'] + list_deltas[4]['O2']
#dataframe['ratio'] = list_ratios[2]['C4'] + list_ratios[2]['P4'] + list_ratios[2]['O2'] + list_ratios[3]['C4'] + list_ratios[3]['P4'] + list_ratios[3]['O2'] + list_ratios[4]['C4'] + list_ratios[4]['P4'] + list_ratios[4]['O2']

electrodes = []
for i in range(2,5):
  electrodes += (['C4']*len(list_alphas[i]['C4']) + ['P4']*len(list_alphas[i]['P4']) + ['O2']*len(list_alphas[i]['O2'])  + ['Fp']*len(list_alphas[i]['Fp']))

dataframe['electrode'] = electrodes
```

```
import statsmodels.api as sm
from statsmodels.formula.api import ols

for frequency in ['baratio', 'alpha', 'beta', 'theta']:
  test = frequency+'~ State'
  for channel in ['Fp', 'O2', 'C4', 'P4']:
    mod = ols(test, data=dataframe[(dataframe['electrode'] == channel)]).fit()
                  
    aov_table = sm.stats.anova_lm(mod, typ=2)
    print('result for {} for electrode {}'.format(frequency, channel))
    print(aov_table)
    pair_t = mod.t_test_pairwise('State')
    print(pair_t.result_frame)
```
